```python
from bokeh.plotting import figure, output_file, show
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
output_file('line.html')
fig= figure(title= 'Line plot', x_axis_label='x', y_axis_label='y')
fig.line(x, y)
show(fig)
```

```python
from bokeh.plotting import figure, output_file, show

# Create a new figure
fig = figure(width=800, height=200)

# Add horizontal bars to the figure
fig.hbar(y=[2, 4, 6], height=1, left=0, right=[1, 2, 3], color='cyan')

# Specify the output file
output_file('bar.html')

# Display the figure
show(fig)

```

```python
from bokeh.plotting import figure, output_file, show
fig = figure(width = 800, height = 200)
fig.vbar(x=[2,4,6], width = 1, bottom = 0, top = [10, 8, 4], color = 'Cyan' )
output_file('vbar.html')
show(fig)
```

```python
from bokeh.plotting import figure, output_file, show
p=figure(width = 800, height = 400)
p.scatter([1, 4, 3, 2, 5], [6, 5, 2, 4, 7], marker = 'circle', size = 20, fill_color='Green')
output_file('scatter.html')
show(p)
```

```python
from bokeh.plotting import figure , output_file, show
from math import pi
p=figure(width = 400, height = 400)
p.rect(x=[1,2, 3], y=[1,2,3], width=0.2, height=40, color='Green', angle=pi/3,height_units='screen')
output_file('Rectangle.html')
show(p)
```

```python
from bokeh.plotting import figure, output_file, show
p=figure(width = 800, height = 400)
x = [1, 2, 5, 7, 9]
y1= [2, 5, 4, 6, 8]
y2= [5, 9, 11, 12, 15]
p.varea(x=x, y1=y1,y2=y2)
output_file('area.html')
show(p)
```

```python
import plotly.graph_objects as go
import numpy as np
# Helix equation
t = np.linspace(0, 10, 50)
x, y, z = np.cos(t), np.sin(t), t
fig = go.Figure(data=[go.Scatter3d(x=x, y=y,z=z, mode='markers')])
fig.show()
```

```python
import pandas as pd
import plotly.express as px
#import data
data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2011_us_ag_exports.csv')
# create choropleth map for the data
# color will be the column to be color-coded
# locations is the column with sppatial coordinates
fig = px.choropleth(data, locations='code',locationmode="USA-states",color='total exports',scope="usa")
fig.show()
```
